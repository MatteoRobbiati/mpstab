```python
from qibo.ui.mpldrawer import plot_circuit

from tncdr.targets.ansatze import (
    HardwareEfficient,
    TranspiledAnsatz,
)
```

```python
nqubits = 10
nlayers = 1
```

```python
theoretical_ansatz = HardwareEfficient(nqubits=nqubits, nlayers=nlayers)
theoretical_circuit = theoretical_ansatz.circuit
```

```python
_, fig = plot_circuit(theoretical_circuit, style={"dpi": 120}, cluster_gates=True)
```

```python
TranspiledAnsatz
```

```python
transpiled_ansatz = TranspiledAnsatz(original_circuit=theoretical_circuit)
transpiled_circuit = transpiled_ansatz.circuit
```

```python
_, fig = plot_circuit(transpiled_circuit, style={"dpi": 120}, cluster_gates=True)
```

```python
partitionated_circuit, magic_layers, stabilizer_layers = transpiled_ansatz.partitionate_circuit(replacement_probability=0.5)
```

```python
len(magic_layers), len(stabilizer_layers)
```

```python
for lay in (stabilizer_layers):
    lay.draw()
    print("\n")
```
